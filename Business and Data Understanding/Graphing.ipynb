```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import string
```

```python
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
plt.rcParams["figure.figsize"] = [10, 6]
%config InlineBackend.figure_format = 'retina'
```

```python
# scatter plot for the runtime and profit for the top 6 most profitable genres

fig, ax = plt.subplots(figsize = (15,7))
sns.scatterplot(x='sqft_living',y='price', data=df, ax=ax)
ax.set_title('sqft_living and Price', fontsize=20)
ax.set_xlabel('sqft_living', fontsize=15)
ax.set_ylabel('Price', fontsize=15)
#plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
fig.tight_layout();            
# recommod movie that, it should run from 75 to 145 mintues.
```

```python
# here we are creating a bar chart to compare the genres with the profit of each movie (world_profit)
fig, ax = plt.subplots(figsize = (18,12))
sns.scatterplot(x = 'price', y = 'sqft_above', data = df,  ax = ax)
ax.set_title('Sqft above and Price', fontsize=30)
ax.set_ylabel('Sqft Above', fontsize=22)
ax.set_xlabel('Price in M', fontsize=22)
fig.tight_layout();
```

```python
# here we are creating a bar chart to compare the genres with the profit of each movie (world_profit)
fig, ax = plt.subplots(figsize = (20,12))
sns.barplot(x = 'bedrooms', y = 'sqft_living', data = df,  ax = ax)
ax.set_title('Bedrooms and Living Square Foot', fontsize=30)
ax.set_ylabel('Price in M', fontsize=22)
ax.set_xlabel('Bedrooms', fontsize=22)
fig.tight_layout();
```

```python
# here we are creating a bar chart to compare the genres with the profit of each movie (world_profit)
fig, ax = plt.subplots(figsize = (20,12))
sns.barplot(x = 'bathrooms', y = 'price', data = df,  ax = ax)
ax.set_title('Bathrooms and Price', fontsize=30)
ax.set_ylabel('Price in M', fontsize=22)
ax.set_xlabel('Bathrooms', fontsize=22)
fig.tight_layout();
```
